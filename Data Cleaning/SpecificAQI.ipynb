```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandasql as ps
import glob
from dateutil import parser

import warnings
warnings.filterwarnings('ignore')
```

```python
PM25_88101_raw = pd.read_csv("data/Gases&PM_AQI/PM2.5_88101_2016.csv")
PM25_88502_raw = pd.read_csv("data/Gases&PM_AQI/PM2.5_88502_2016.csv")
CO_raw = pd.read_csv("data/Gases&PM_AQI/CO_42101_2016.csv")
NO2_raw = pd.read_csv("data/Gases&PM_AQI/NO2_42602_2016.csv")
Ozone_raw = pd.read_csv("data/Gases&PM_AQI/Ozone_44201_2016.csv")
PM10_raw = pd.read_csv("data/Gases&PM_AQI/PM10_81102_2016.csv")
SO2_raw = pd.read_csv("data/Gases&PM_AQI/SO2_42401_2016.csv")
```

```python
PM25_88101 = PM25_88101_raw[['State Code', 'County Code','Site Num','Date Local', 'AQI','State Name', 'County Name']]
PM25_88101.rename({'AQI':'AQI_PM2.5_88101'}, axis=1, inplace=True)
PM25_88101.rename({'State Name':'State', 'County Name': 'County'}, axis=1, inplace=True)
PM25_88101['Date Local'] = pd.to_datetime(PM25_88101['Date Local'])
PM25_88101['Year'] = pd.DatetimeIndex(PM25_88101['Date Local']).year
PM25_88101['Month'] = pd.DatetimeIndex(PM25_88101['Date Local']).month
PM25_88101['Day'] = pd.DatetimeIndex(PM25_88101['Date Local']).day
PM25_88101.drop('Date Local', axis=1, inplace=True)
PM25_88101.dropna(axis = 0, subset=['AQI_PM2.5_88101'], inplace=True)
PM25_88101.head()

PM25_88502 = PM25_88502_raw[['State Code', 'County Code','Site Num','Date Local', 'AQI','State Name', 'County Name']]
PM25_88502.rename({'AQI':'AQI_PM2.5_88502'}, axis=1, inplace=True)
PM25_88502.rename({'State Name':'State', 'County Name': 'County'}, axis=1, inplace=True)
PM25_88502['Date Local'] = pd.to_datetime(PM25_88502['Date Local'])
PM25_88502['Year'] = pd.DatetimeIndex(PM25_88502['Date Local']).year
PM25_88502['Month'] = pd.DatetimeIndex(PM25_88502['Date Local']).month
PM25_88502['Day'] = pd.DatetimeIndex(PM25_88502['Date Local']).day
PM25_88502.drop('Date Local', axis=1, inplace=True)
PM25_88502.dropna(axis = 0, subset=['AQI_PM2.5_88502'], inplace=True)
PM25_88502.head()

CO = CO_raw[['State Code', 'County Code','Site Num','Date Local', 'AQI','State Name', 'County Name']]
CO.rename({'AQI':'AQI_CO'}, axis=1, inplace=True)
CO.rename({'State Name':'State', 'County Name': 'County'}, axis=1, inplace=True)
CO['Date Local'] = pd.to_datetime(CO['Date Local'])
CO['Year'] = pd.DatetimeIndex(CO['Date Local']).year
CO['Month'] = pd.DatetimeIndex(CO['Date Local']).month
CO['Day'] = pd.DatetimeIndex(CO['Date Local']).day
CO.drop('Date Local', axis=1, inplace=True)
CO.dropna(axis = 0, subset=['AQI_CO'], inplace=True)
CO.head()

NO2 = NO2_raw[['State Code', 'County Code','Site Num','Date Local', 'AQI','State Name', 'County Name']]
NO2.rename({'AQI':'AQI_NO2'}, axis=1, inplace=True)
NO2.rename({'State Name':'State', 'County Name': 'County'}, axis=1, inplace=True)
NO2['Date Local'] = pd.to_datetime(NO2['Date Local'])
NO2['Year'] = pd.DatetimeIndex(NO2['Date Local']).year
NO2['Month'] = pd.DatetimeIndex(NO2['Date Local']).month
NO2['Day'] = pd.DatetimeIndex(NO2['Date Local']).day
NO2.drop('Date Local', axis=1, inplace=True)
NO2.dropna(axis = 0, subset=['AQI_NO2'], inplace=True)
NO2.head()

SO2 = SO2_raw[['State Code', 'County Code','Site Num','Date Local', 'AQI','State Name', 'County Name']]
SO2.rename({'AQI':'AQI_SO2'}, axis=1, inplace=True)
SO2.rename({'State Name':'State', 'County Name': 'County'}, axis=1, inplace=True)
SO2['Date Local'] = pd.to_datetime(SO2['Date Local'])
SO2['Year'] = pd.DatetimeIndex(SO2['Date Local']).year
SO2['Month'] = pd.DatetimeIndex(SO2['Date Local']).month
SO2['Day'] = pd.DatetimeIndex(SO2['Date Local']).day
SO2.drop('Date Local', axis=1, inplace=True)
SO2.dropna(axis = 0, subset=['AQI_SO2'], inplace=True)
SO2.head()

Ozone = Ozone_raw[['State Code', 'County Code','Site Num','Date Local', 'AQI','State Name', 'County Name']]
Ozone.rename({'AQI':'AQI_Ozone'}, axis=1, inplace=True)
Ozone.rename({'State Name':'State', 'County Name': 'County'}, axis=1, inplace=True)
Ozone['Date Local'] = pd.to_datetime(Ozone['Date Local'])
Ozone['Year'] = pd.DatetimeIndex(Ozone['Date Local']).year
Ozone['Month'] = pd.DatetimeIndex(Ozone['Date Local']).month
Ozone['Day'] = pd.DatetimeIndex(Ozone['Date Local']).day
Ozone.drop('Date Local', axis=1, inplace=True)
Ozone.dropna(axis = 0, subset=['AQI_Ozone'], inplace=True)
Ozone.head()

PM10 = PM10_raw[['State Code', 'County Code','Site Num','Date Local', 'AQI','State Name', 'County Name']]
PM10.rename({'AQI':'AQI_PM10'}, axis=1, inplace=True)
PM10.rename({'State Name':'State', 'County Name': 'County'}, axis=1, inplace=True)
PM10['Date Local'] = pd.to_datetime(PM10['Date Local'])
PM10['Year'] = pd.DatetimeIndex(PM10['Date Local']).year
PM10['Month'] = pd.DatetimeIndex(PM10['Date Local']).month
PM10['Day'] = pd.DatetimeIndex(PM10['Date Local']).day
PM10.drop('Date Local', axis=1, inplace=True)
PM10.dropna(axis = 0, subset=['AQI_PM10'], inplace=True)
PM10.head()
```

```python
PM25 = pd.merge(PM25_88101, PM25_88502, how="outer",on=["Year", "State", "County", "Month", "Day","State Code","County Code", "Site Num"])
PM25['AQI_PM2.5'] = PM25[["AQI_PM2.5_88101", "AQI_PM2.5_88502"]].max(axis=1)
PM25.head()
```

```python
specificAQI = pd.merge(PM25, PM10, how='outer', on=["Year", "State", "County", "Month", "Day","State Code","County Code", "Site Num"])
specificAQI = pd.merge(specificAQI, CO, how='outer', on=["Year", "State", "County", "Month", "Day","State Code","County Code", "Site Num"])
specificAQI = pd.merge(specificAQI, NO2, how='outer', on=["Year", "State", "County", "Month", "Day","State Code","County Code", "Site Num"])
specificAQI = pd.merge(specificAQI, Ozone, how='outer', on=["Year", "State", "County", "Month", "Day","State Code","County Code", "Site Num"])
specificAQI = pd.merge(specificAQI, SO2, how='outer', on=["Year", "State", "County", "Month", "Day","State Code","County Code", "Site Num"])
specificAQI['State'] = specificAQI['State'].str.lower()
specificAQI['County'] = specificAQI['County'].str.lower()
specificAQI.head()
```
