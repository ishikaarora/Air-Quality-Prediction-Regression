# Merge Data

```python
merged = pd.merge(aqi, population, on=['State', 'County','Year'], how='left')
merged.drop(['StateNum','CountyNum','Region'], axis=1, inplace=True)
merged = pd.merge(merged, powerplant, on=['State', 'County','Year','Month'], how='left')
merged = pd.merge(merged, temps, on=['State', 'County','Year','Month'], how='left')
display(merged[10000:10200])
```
